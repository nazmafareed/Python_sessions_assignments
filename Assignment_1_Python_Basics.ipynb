###**Question#1. Variable**
 **What is a variable in Python? Write the syntax to create a variable and give one example.**

###**Answer:**
 Variables are used to store data in a program.They act as containers for values that can be changed or manipulated during the execution of a program.
####**Syntax:**
variable_name = value  
**Example:**  
age =18 
name="Nazma"

###**Question#2.Rules of Variable Naming**
**Write any three rules for naming variables in Python.Give one correct and one incorrect example.**
###**Answer:**
In python, there are specific rules we must follow when naming a variable:

  **1. Start with a Letter or Underscore:** A variable must start with a letter (A-Z) or underscore (_).It cannot start with a number.

**2. No special characters:** Only alpha-numeric characters and underscores (A-z,0-9,and _) are allowed.

 **3. Case-Senstivity:** Variables names are case-sensitive (e.g.,'age' and 'Age' are different).  
 ###### **Examples of Variable Naming:**
 * ***Correct Example:***  
 user_name = "Zunaira"  
  ###### (Explanation:It starts with a letter and contains no special characters except underscore.)
 * ***Incorrect Example:***  
 1st_@user ="Zohan"  
 ###### (Explanation:This is incorrect because a variable name cannot start with a number "1" and contains a special character "@",both of which are not allowed in Python variables names.)

###**Question#9:Type Casting**
**What is type casting?Write a Python example to convert a string into an integer.**
###**Answer:**
Type Casting is the process of converting a value from one data type to another.For example, converting a string "10" into mathematical integer 10.
* **Example:**  
 Using the int() function to convert a string into an integer

```python
# Question 9: Converting string to integer
# This is called Type Casting

string_number = "500"          # A number inside quotes is a string
converted_number = int(string_number) # Converting string to int

# Printing the result
print(converted_number)
print(type(converted_number))  # This will now show <class 'int'>
```
